```python
class GunDetection:
    
    def __init__(self, modelpath, camerapath):
        self.trained_model = YOLO(Path(modelpath))
        self.camerapath = camerapath

    def gundetection(self):
        videoframe = cv2.VideoCapture(str(self.camerapath))
        while True:
            frame, pixel = videoframe.read()
            if frame == True:         
                resized_image = cv2.resize(pixel, (1280, 720))
                #resized_image = cv2.resize(pixel, (640, 640))
                resized_image = cv2.convertScaleAbs(resized_image, alpha=1.5, beta=40)
                #resized_image = pixel
                prediction = self.trained_model.predict(resized_image, conf = 0.30)[0].boxes
                bb = prediction.xyxy.numpy().astype(int) 
                clsid = prediction.cls.numpy().astype(int)
                for (x1, y1, x2, y2), ids in zip(bb, clsid):
                    label = self.trained_model.names[ids]
                    x_center = (x1+x2)/2
                    y_center = (y1+y2)/2
                    cv2.rectangle(resized_image, (int(x1),int(y1)), (int(x2),int(y2)), (0, 0, 255), 3)
                    cv2.putText(resized_image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 3)
                cv2.imshow("Video", resized_image)
                key = cv2.waitKey(1)
                if key == 13:
                    break
        cv2.destroyAllWindows()
        videoframe.release()

detection1 = GunDetection(Path(r"C:\Users\amirp\Documents\My Projects\Project_1_Gun_Detection\Trained_Models\last.pt"),
                          Path(r"C:\Users\amirp\Documents\My Projects\Project_1_Gun_Detection\Videos\Final_Video_1.mp4"))
detection1.gundetection()
```
